```python
import matplotlib.pyplot as plt

# Data for plotting
xAxis = list(range(1,13))
yAxis = [10,20,25,30,39,3,54,84,88,99,7,45]

# create the plot plt.plot(xAxis,yAxis)
plt.plot(xAxis,yAxis,marker = "o",color="green",linestyle="dashed",markerfacecolor="black",markersize=10)

#Labeling the axes and the plot
plt.plot(xAxis,label = "Days")
plt.plot(xAxis,yAxis,label = "Temperature in F")
plt.legend(loc= "upper left") # Places the lengend at the upper lleft

plt.title("Temperature line graph")
plt.show()
```

```python
flowers=['rose','lily','sunflower','marigold','lavender','jasmine']
count=[10,25,40,15,65,5]
plt.xlabel('quantity of flowers')
plt.ylabel('flowers')
plt.title('Flowers quantity sale')
plt.bar(flowers,count,color="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt

fruits = ["Apples", "Bananas", "Cherries", "Grapes", "Oranges"]
votes = [15, 30, 25, 10, 20]


plt.figure(figsize=(8, 5))
plt.bar(fruits, votes, color=['red', 'yellow', 'pink', 'purple', 'orange'])
plt.xlabel("Fruits")
plt.ylabel("Number of Votes")
plt.title("Favorite Fruits Poll")

plt.show()
```

```python
#Horizontal Bar graph 
ages=["10-20","25-30","35-40","40-50","60-70"]
populationCount=[10000,40000,20000,450000,60000]
plt.barh(ages,populationCount,color="white",edgecolor="red")
plt.xlabel("populationCount")
plt.ylabel("ages")
plt.title("Population by age")
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,shadow="True",autopct='%1.2f', radius=1,explode=[0,0,0,0.1,0.2],counterclock=True,startangle=45)
plt.axis("equal")
plt.show()
```

```python
#scatter plot

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
sizes = [100, 200, 300, 400, 500]  
colors = ['red', 'green', 'blue', 'orange', 'purple'] 

plt.scatter(x, y, s=sizes, c=colors,alpha=0.1)

plt.xlabel('Values')
plt.ylabel('Numbers')
plt.title('Scatter Plot')

plt.show()
```
